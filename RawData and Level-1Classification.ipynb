```python
import os
import numpy as np
import cv2
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
ptodata="./mpimages/"
ptocdata="./cropped/"
```

```python
img_dir=[]
for i in os.scandir(ptodata):
    if i.is_dir():
        img_dir.append(i.path)
```

```python
img_dir
```

```python
import shutil
if os.path.exists(ptocdata):
     shutil.rmtree(ptocdata)
os.mkdir(ptocdata)
```

```python
face_cascade = cv2.CascadeClassifier(r"C:\Users\A\anaconda3\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier(r"C:\Users\A\anaconda3\Lib\site-packages\cv2\data\haarcascade_eye.xml")
```

```python
def get_cropped_img(imgpath):
    img=cv2.imread(imgpath)
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,1.3,5)
    for(x,y,w,h) in faces:
        roi_gray=gray[y:y+h,x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            return roi_color
    
```

```python
cimg_dir=[]
file_names_dict={}

for i in img_dir:
    count=1
    filename=i.split('/')[-1]
    file_names_dict[filename] = []
    for j in os.scandir(i):
        roi_color=get_cropped_img(j.path)
        if roi_color is not None:
            cfolder=ptocdata+filename
            if not os.path.exists(cfolder):
                os.makedirs(cfolder)
                cimg_dir.append(cfolder)
                print("Generating folder: ",cfolder)
            cropped_file_name = filename + str(count) + ".jpg"
            cropped_file_path = cfolder + "/" + cropped_file_name 
            
            cv2.imwrite(cropped_file_path,roi_color)
            file_names_dict[filename].append(cropped_file_path)
            count += 1
```
